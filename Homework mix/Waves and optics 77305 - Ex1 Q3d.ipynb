```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def A(a, b):
  return np.sqrt(a**2 + b**2)

def Phi(a, b):
  return np.arctan(b/a)
```

```python
# constants
a = 1.0 # [Hz]
w0 = 1.0 # [rad/sec]

# frequencies vector
w = np.linspace(-10, 10, 100, endpoint=True)

# real and imaginary parts of FT{f}
real = (-a)/(a**2 + (w0-w)**2)
imaginary = -(w0-w)/(a**2 + (w0-w)**2)

# plot
plt.figure()
plt.plot(w, A(real, imaginary), label='A(w)')
plt.plot(w, Phi(real, imaginary), label='Phi(w)')
plt.legend()
plt.grid()
plt.axis([-10, 10, -1.5, 1.5])
plt.show()
```
