```python
import matplotlib as mpl
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp
from scipy.special import jv, factorial, gamma, jn_zeros, yn, yn_zeros
```

```python
plt.style.use(['science', 'notebook', 'grid'])
plt.rcParams["figure.figsize"] = (8,4)
```

```python
N = 6
x = np.linspace(-N*np.pi, N*np.pi, 200)
plt.plot(x, np.sin(x), label='sin(x)')
zeroes = [x*np.pi for x in range(-N, N+1, 1)]
plt.scatter(zeroes, np.sin(zeroes), label='sin(x)=0')
plt.title('Kai Carmin')
plt.legend(loc='upper right')
```

```python
plt.figure()
x = np.linspace(-N*np.pi,N*np.pi, 200)

for i in range(0,3):
    plt.plot(x, jv(i, x), label=f'order {i}')
plt.legend()
plt.title('Bessel function of the first kind (Kai Carmin)')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
n=3
s=0

plt.figure()
y=0
x = np.linspace(0,1, 100)
for m in range(n+1):
    y += ((((-1)**m) * ((x/2)**(2*m+s)))/(factorial(m) * gamma(m+s+1)))
    plt.plot(x, y, label=f'm={m}')
plt.legend()
plt.title('Bessel function 1st kind, order 0, 4 first terms summed (Kai Carmin)')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
N = 10

plt.figure()
x = np.linspace(-N*np.pi,N*np.pi, 800)

for i in range(0,3):
    plt.plot(x, yn(i, x), label=f'order {i}')
plt.scatter(yn_zeros(0, 10), np.zeros([1,10]), label=r'$k_{0m}$')
plt.legend()
plt.title('Bessel function of the second kind (Kai Carmin)')
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-3, 1)
```

```python
def H_k(k):
    h = 0
    for i in range(0, k+1):
        h += 1/k
    return h
```

```python
n=3
s=0

plt.figure()
x = np.linspace(0,1, 100)
y = (2/np.pi)*((np.log(x/2)+np.euler_gamma)*jv(0, x))
for k in range(0, n+1):
    if k>0:
        y += (2/np.pi)*(((-1)**(k+1))*H_k(k)*((0.25*x**2)**k)/(factorial(k)**2))
    plt.plot(x, y, label=f'm={k}')
plt.legend()
plt.title('Bessel function 1st kind, order 0, 4 first terms summed (Kai Carmin)')
plt.xlabel('X')
plt.ylabel('Y')
```
